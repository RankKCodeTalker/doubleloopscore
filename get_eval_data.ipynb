转换成这种格式：
```json
[
    {
        "question": "",
        "answers": [
            {
                "answer": "",
                "score": 0
            },
            {
                "answer": "",
                "score": 0
            }
        ]
    }
]
```

````python
import pandas as pd
from myutils import io
data = pd.read_excel("data/evalv2.xlsx")
data
````

````python
data.columns
````

````python
columns_answers = [
    ('0 answer', 0),
    ('25 answer', 25),
    ('50 answer', 50),
    ('75 answer', 75),
    ('100 answer', 100)
]
result = []
for i in range(len(data)):
    tmp = {
        "question": data.loc[i, 'query'],
        "answers": []
    }
    for c1, c2 in columns_answers:
        tmp['answers'].append({
            "answer": data.loc[i, c1] if not pd.isna(data.loc[i, c1]) else "",
            "score": c2
        })
    result.append(tmp)
io.jsondump(result, "data/eval.json")
````
